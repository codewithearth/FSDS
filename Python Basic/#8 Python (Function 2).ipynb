```python
def getfib(n):
    a = 0
    b = 1
    output = []
    for i in range(n):
        output.append(a)
        a,b = b,a+b
    return output
```

```python
getfib(10)
```

```python
def getfib1(n):
    a = 0
    b = 1
    for i in range(n):
        yield a
        a,b = b,a+b
```

```python
getfib1(20000)
```

```python
for i in getfib(500):
    print(i)
```
